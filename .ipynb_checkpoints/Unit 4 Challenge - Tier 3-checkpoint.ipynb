**2.6. Visualizing the data**

To visualize the data, why not subset on a particular London Borough? Maybe do a line plot of Month against Average Price?

```python
import random

def plot_borough(borough):
    dates = list(properties.columns)
    plt.xlabel('Year')
    plt.ylabel('Average Price')
    prices = list(properties.loc[borough])
    plt.plot( dates, prices)
    plt.title(f'Average Prices of Properties in {borough}')
    plt.margins(x=0)
    plt.show()


# Using a random London Borough
random_borough = properties.index[random.randint(0, 31)]
plot_borough(random_borough)

# Following code is for all boroughs.
def plot_all(table):
    plt.xlabel('Year')
    plt.ylabel('Average Price')
    plt.title(f'Average Price vs Year')
    for borough in table.index:
        dates = list(table.columns)
        prices = list(table.loc[borough])
        plt.plot(dates, prices, label = f'{borough}')
        plt.legend(bbox_to_anchor=(-.2, 1.1))
    plt.ticklabel_format(style='plain', useOffset=False, axis='y')
    plt.margins(x=0)
    plt.show()
    
    
plot_all(properties)
```

**3. Modeling**

Consider creating a function that will calculate a ratio of house prices, comparing the price of a house in 2018 to the price in 1998.

Consider calling this function create_price_ratio.

You'd want this function to:
1. Take a filter of dfg, specifically where this filter constrains the London_Borough, as an argument. For example, one admissible argument should be: dfg[dfg['London_Borough']=='Camden'].
2. Get the Average Price for that Borough, for the years 1998 and 2018.
4. Calculate the ratio of the Average Price for 1998 divided by the Average Price for 2018.
5. Return that ratio.

Once you've written this function, you ultimately want to use it to iterate through all the unique London_Boroughs and work out the ratio capturing the difference of house prices between 1998 and 2018.

Bear in mind: you don't have to write a function like this if you don't want to. If you can solve the brief otherwise, then great! 

***Hint***: This section should test the skills you acquired in:

```python
# Creating a properties by year table:
# Takes year as INT
def extract_year(year):
    year_start = properties1.columns >= np.datetime64(pd.to_datetime(year,format='%Y'))
    year_end = properties1.columns < np.datetime64(pd.to_datetime(year+1,format='%Y'))
    return properties1.loc[:,year_start & year_end]

properties_by_year = pd.DataFrame()
properties_by_year.index = properties1.index

for year in range(1995, 2024):
    col = extract_year(year).mean(axis=1)
    properties_by_year[year] = col

# Create price ratio, but why do it for one borough when you can easily implement it for all boroughs?
def create_price_ratio(end, start):
    price_end = properties_by_year.loc[:, end]
    price_start = properties_by_year.loc[:, start]
    return (price_end/price_start)

solution1 = create_price_ratio(2018, 1998)
solution1 = solution1.sort_values(ascending=False)
solution1 = pd.DataFrame(solution1.head(10))
solution1 = solution1.rename(columns={0: "Average Annual Price Increase Ratio of 2018/1998."})
solution1
```

```python
# Returns absolute average price difference over 2 specified years
def create_price_difference(end, start):
    price_end = properties_by_year.loc[:, end]
    price_start = properties_by_year.loc[:, start]
    return (price_end-price_start)

average_price_increase = create_price_difference(2018, 1998)
average_price_increase = average_price_increase.sort_values(ascending=False)
average_price_increase = pd.DataFrame(average_price_increase.head(10))
average_price_increase = average_price_increase.rename(columns={0: "Average Annual Price Increase from 1998 to 2018."})
average_price_increase
```

```python
# The challenge where my mentor suggested to do the above using groupby:
properties2 = properties2.transpose()
properties2 = properties2.reset_index()
properties2 = properties2.rename(columns={'Unnamed: 0': 'dates'})
properties2 = properties2.groupby(properties2.dates.dt.year)[properties2.columns[1:]].mean()
solution2 = pd.DataFrame((properties2.loc[2018]/properties2.loc[1998]).sort_values(ascending=False))
solution2 = solution2.rename(columns={0: "Average Annual Price Increase Ratio of 2018/1998."})
solution2.head(10)
```
